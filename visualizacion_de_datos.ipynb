Visualizacion de datos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Ejemplo 1
X = ['a', 'b', 'v', 'e', 'd']

# Creamos dos series de datos
datos_1 = [10,20,30,40,50]
datos_2 = [15,25,35,37,65]
```

```python
# Creamos el lienzo
plt.plot(X, datos_1)
plt.plot(X, datos_2)
plt.show()
```

```python
# Ejemplo 2
# Cree una grafica para mostrar las entregas y devoluciones de envios de enero a junio, ademas agregar un eje secundario
# para mostrar los porcentajes de entrega y la meta de cada mes.

mes = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']
entregas = np.random.randint(300, 1200, size=6)
devoluciones = np.random.randint(50, 150, size=6)
meta = [0.85, 0.85, 0.85, 0.85, 0.85, 0.85]

porcentajes = []
for i in range(len(mes)):
    porcentage_entregas = entregas[i] / (entregas[i] + devoluciones[i])
    porcentajes.append(porcentage_entregas)


rango_mes = np.arange(len(mes))
ancho_barra = 0.4

print('Entregas: ', entregas)
print('Devoluciones', devoluciones)
print('Porcentaje', porcentajes)
print('Porcentaje', meta)
```

```python
fig, axi1 = plt.subplots();

# Graficas de barras combinadas
axi1.bar(rango_mes - ancho_barra/2, entregas, width=ancho_barra, label='Entregas', color='#2d2d2d')
axi1.bar(rango_mes + ancho_barra/2, devoluciones, width=ancho_barra, label='Devoluciones', color='red')

# Mostramos las leyendas
plt.legend()
# Mostrar un titulo
plt.title('Entregas y devoluciones del primer semestre del año 2025')
# Etiquetas de los ejes
plt.xlabel('Mes')
plt.ylabel('Envios')

# Grafica de linea en porcentajes
axi2 = axi1.twinx();
axi2.plot(mes, porcentajes, color='green', marker='o', label="% Entregas")
# Grafica de linea de la meta
axi2.plot(mes, meta, color='blue', marker='>')

# # Mostramos una escala mas personalizada de 0% a 100% en pasos de 10%
axi2.set_yticks(np.arange(0, 1.1, 0.1))
axi2.set_ylim(0, 1)  # Asegurar que el eje va de 0 a 1

# Para mostrar un grid
plt.grid()
# Mostramos todo el grafico
plt.show()
```
