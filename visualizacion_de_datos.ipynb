Visualizacion de datos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Ejemplo 1
X = ['a', 'b', 'v', 'e', 'd']

# Creamos dos series de datos
datos_1 = [10,20,30,40,50]
datos_2 = [15,25,35,37,65]
```

```python
# Creamos el lienzo
plt.plot(X, datos_1)
plt.plot(X, datos_2)
plt.show()
```

```python
# Ejemplo 2

mes = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']
entregas = np.random.randint(300, 1200, size=6)
devoluciones = np.random.randint(50, 150, size=6)

porcentajes = []
for i in range(len(mes)):
    porcentage_entregas = entregas[i] / (entregas[i] + devoluciones[i]) * 100
    porcentajes.append(porcentage_entregas)


rango_mes = np.arange(len(mes))
ancho_barra = 0.4

print('Entregas: ', entregas)
print('Devoluciones', devoluciones)
print('Porcentaje', porcentajes)
```

```python
# Graficas de barras
plt.bar(rango_mes - ancho_barra/2, entregas, width=ancho_barra, label='Entregas', color='#2d2d2d')
plt.bar(rango_mes + ancho_barra/2, devoluciones, width=ancho_barra, label='Devoluciones', color='red')
# Grafica de linea
plt.plot(mes, porcentajes, color='green', marker='o')
# Mostramos las leyendas
plt.legend()
# Mostrar un titulo
plt.title('Entregas y devoluciones del primer semestre del año 2025')
# Etiquetas de los ejes
plt.xlabel('Mes')
plt.ylabel('Envios')
# Mostramos en el eje y una escala mas personalizada, de 0 a 11 en rango de 1. -> plt.yticks(np.arange(0, 11, 1))
# Para mostrar un grid
plt.grid()
# Mostramos todo el grafico
plt.show()
```
